```python
df["km_driven"]=(df['km_driven'].astype(str).str.replace("kms",'').str.replace('km','').str.replace(',',''))
df['km_driven'] =pd.to_numeric(df['km_driven'])
df.head(10)
```

```python
df['mileage']=(df['mileage'].astype(str).str.replace('kmpl','').str.strip())
df['mileage'] = pd.to_numeric(df['mileage'],errors='coerce')
df.head(10)
```

```python
df = df.drop(columns=['Unnamed: 0','car_name','model'])
df.head(5)
```

```python
df['fuel_type'].unique()
```

```python
df['fuel_type'] =df['fuel_type'].str.strip().str.title()
df['fuel_type'].unique()
```

```python
df['seller_type']=df['seller_type'].str.strip().str.title()
df["seller_type"].unique()
```

```python
for col in ["engine",'mileage','max_power']:
  median_val =df[col].median()
  df[col]=df[col].fillna(median_val)
  print(f"filled{col}with median:{median_val} ")
```

```python
seats_mode =df['seats'].mode()[0]
df['seats'] = df["seats"].fillna(seats_mode)
```

Task 3 : Base line of MAE

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,cross_val_score,KFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
import warnings

print("all  imports ready")
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size =0.2,random_state = 42
)

print(f" train: {x_train.shape[0]} rows ({x_train.shape[0]/len(x)*100:.0f}%)")
print(f" test : {x_test.shape[0]} rows ({x_test.shape[0]/len(x)*100:.0f})")
```

```python
numeric_cols =["vehicle_age","km_driven","mileage","engine","max_power","seats"]

scaler = StandardScaler()
x_train[numeric_cols] = scaler.fit_transform(x_train[numeric_cols])
```

```python
x_train
```

```python
x_test[numeric_cols] =scaler.fit_transform(x_test[numeric_cols])
```

```python
x_test
```

```python
model = LinearRegression()
model.fit(x_train,y_train)

print(f"model trained")
print(f"number of coeffients :{len(model.coef_)}")
print(f"intercept : {model.intercept_:,.0f}")
```

```python
model.coef_
```

```python
y_train_predict = model.predict(x_train)
y_train_predict
```

```python
print(y_train)
```

```python
train_mae = mean_absolute_error(y_train,y_train_predict)
print(train_mae)
```

```python
y_test_predict = model.predict(x_test)
print(y_test_predict)
```

```python
test_mae = mean_absolute_error(y_test,y_test_predict)
print(test_mae)
```
